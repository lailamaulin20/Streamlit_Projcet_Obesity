````python
# Fungsi untuk hapus outlier berbasis IQR
def remove_outliers_iqr(dataframe, col):
    Q1 = dataframe[col].quantile(0.25)
    Q3 = dataframe[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return dataframe[(dataframe[col] >= lower) & (dataframe[col] <= upper)]

# Bersihkan outlier untuk semua fitur numerik
for col in num_cols:
    df = remove_outliers_iqr(df, col)

# Lihat jumlah data setelah cleaning
print("Jumlah data setelah outlier dibuang:", df.shape)
````

# **Preprocessing**

---



````python
from imblearn.over_sampling import SMOTE
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
import pandas as pd

# Encode target label
df_smote = df.copy()
le = LabelEncoder()
df_smote['NObeyesdad'] = le.fit_transform(df_smote['NObeyesdad'])

# One-hot encoding fitur kategorikal
df_encoded = pd.get_dummies(df_smote.drop(columns='NObeyesdad'))
target = df_smote['NObeyesdad']

# SMOTE
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(df_encoded, target)
````

**Tujuan:**

````python
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10]
}
````

**Fungsi:**

- max_depth: Kedalaman maksimum dari pohon. None berarti pohon akan terus berkembang sampai semua daun murni.

- min_samples_split: Jumlah minimum sampel yang dibutuhkan untuk membagi node internal.

**Tujuan:**
Mencari kombinasi parameter terbaik untuk meningkatkan akurasi model Random Forest secara optimal melalui pencarian grid (brute-force).

# **Hypertuning**

---



````python
# Buat model dasar
rf = RandomForestClassifier(random_state=42)

# Inisialisasi GridSearchCV
grid_search = GridSearchCV(estimator=rf, param_grid=param_grid,
                           cv=5, scoring='accuracy', n_jobs=-1, verbose=1)

# Fit ke data latih
grid_search.fit(X_train, y_train)
````

**Fungsi dan Keterangan:**

rf: Inisialisasi model Random Forest dasar sebelum dituning.

- **GridSearchCV:**

- estimator=rf → Model yang akan dituning.

- param_grid=param_grid → Kombinasi parameter yang akan diuji.

- cv=5 → 5-fold cross-validation untuk evaluasi setiap kombinasi.

- scoring='accuracy' → Metrik evaluasi yang digunakan.

- n_jobs=-1 → Gunakan seluruh core CPU untuk percepatan.

- verbose=1 → Menampilkan progress saat proses berjalan.

- **fit():** Menjalankan pencarian grid pada data pelatihan dan menemukan kombinasi parameter terbaik.

**Tujuan:**
Menemukan kombinasi hyperparameter terbaik untuk model RandomForestClassifier agar menghasilkan akurasi tertinggi berdasarkan validasi silang.

````python
# Tampilkan parameter terbaik
print("✅ Best Parameters:", grid_search.best_params_)

# Tampilkan skor akurasi terbaik dari hasil cross-validation
print("📊 Best CV Score (Accuracy):", grid_search.best_score_)
````

**Fungsi dan Keterangan:**
- grid_search.best_params_
➤ Menampilkan kombinasi parameter terbaik (optimal) yang ditemukan oleh GridSearchCV.

- grid_search.best_score_
➤ Menampilkan skor akurasi tertinggi dari hasil cross-validation selama pencarian grid.

**Tujuan:**
Untuk mengetahui parameter dan performa terbaik yang diperoleh setelah tuning model RandomForestClassifier, sebelum digunakan untuk prediksi atau evaluasi lebih lanjut.

````python
from sklearn.metrics import classification_report, accuracy_score

# Gunakan model terbaik
best_rf = grid_search.best_estimator_

# Prediksi di data uji
y_pred = best_rf.predict(X_test)

# Evaluasi
print("🎯 Akurasi di Data Uji:", accuracy_score(y_test, y_pred))
print("📋 Classification Report:\n", classification_report(y_test, y_pred))
````
